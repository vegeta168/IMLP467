### 03 -  猜數字（Guess the number）
#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，
#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，
#### 並統計使用者的嘗試次數。如果使用者輸入的數字超出設定範圍，就顯示錯誤提示。

![](data\guess.png)


````python
#設定輸入隨機數工具
import random
#先輸入所需數值
Max = 100
Min = 0
ans = 66
#設定答案所需隨機數範圍
ans = random.randint(1,99) #.這個點是表示開啟random的工具，rand(隨機)int(整數)
#從0開始計算
count = 0
#while是無上限迴圈，if 是知道有範圍的，true的功能只要後面條件是true就無限重覆，直到break
while True:
    guess = int(input(f'請輸入{Min}<?<{Max}範圍的數字:')) #f可讓多個條件併一起
    if guess < Min or guess > Max: #視邏輯or and
        print(f'你看不懂中文嗎？重新輸入！')
        continue
    count += 1 #計數成立後再算次數，所以要放這行
    if guess == ans:
        print('恭喜您，答對了！')
        break #答對就結束
    elif guess < ans:
        Min = guess
        print(f'太小了，您猜了{count}次，再猜一次')
    else:
        Max = guess
        print(f'太大了，您猜了{count}次，再猜一次')
````

### 04 - 計算字數（Word count）
#### 使用者輸入一段文字或讀取一檔案，程式統計字數。

> 輸入(Input):我要成為寫程式的專家 <br>
輸出(Output):你用了 10 個文字述說內心的想法

#### 在特定的文章字串中搜尋輸入的特定字，請使用者輸入欲搜尋的文字，印出總共有幾個。<br>
>輸入(Input):天<br>
輸出(Output): 9

* 進階:計算文章中去除標點符號後的字數，找出文章中出現最多的字與次數

````python
word = input('請輸入你內心的想法：')
print(f'你用了 {len(word)} 個文字述說內心的想法')
query = input('可以查詢內心想法關鍵字出現幾次：')
print(f'關鍵字 【{query}】 一共出現了 {word.count(query)} 次')

s = '''看到了雲，就想起了她的衣裳，
看到了花，就想起了她的容貌，
當春風吹拂著窗櫺、露水正濃的時候。
假若不是在群玉山頭會見過她，
那便是在瑤臺月光下相逢過。'''

# 1. 清理標點符號
for i in '。，\n；：？「」' :
    s = s.replace(i, '')

query = input('可以查詢文章關鍵字出現幾次：')
print(f'關鍵字 【{query}】 一共出現了 {s.count(query)} 次')

# 2. 找出出現最多次的字
maxct = 0      # 從零起算，紀錄最高數字
maxstr = ''    # 記錄那個字是誰

print('文章總字數：', len(s))

for serch in s:
    act = s.count(serch)  
    
    #「比較邏輯」
    if act > maxct:
        maxct = act      # 如果現在這格數字比較大，就更新最高紀錄
        maxstr = serch    # 記下這個字

# 迴圈跑完後，印出結果
print(f"出現最多次的字是：【{maxstr}】，共出現了 {maxct} 次")
#若你最後一行maxct改成act，會變成程序判讀文章最後一個【過】字的次數(每一字巡邏)
````

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```
米花市帝丹小學一年级B班正舉辦期中考試
數學及格的有：柯南、灰原、步美、美環、光彦
英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
m={'柯南','灰原','步美','美環','光彦'} #數學及格
e={'柯南','灰原','丸尾','野口','步美'} #英文及格
h = e | m #全班人員
a = m & e #兩者都及格
b = h - m #數學不及格
c = h - e #英文不及格
d = m - e #數學及格但英文不及格
f = e - m #英文及格但數學不及格
g = e ^ m #有一科及格

print(f'全班             人員{sorted(h)}')
print(f'英數都及格的人：{sorted(a)}')
print(f'數學不及格的人：{sorted(b)}')
print(f'英文不及格的人：{sorted(c)}')
print(f'數學及格但英文不及格：{sorted(d)}')
print(f'英文及格但數學不及格：{sorted(f)}')
print(f'有一科及格：{sorted(g)}')
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
def is_palindrome(text):
    # 1. 清理：只留下字母和數字 (isalnum)，並統一轉小寫
    # 這行取代了原本複雜的 stripAnalphabetics 函式
    clean_text = "".join(char.lower() for char in text if char.isalnum())
    
    # 2. 判斷：正著讀 (clean_text) 是否等於反著讀 (clean_text[::-1])
    return clean_text == clean_text[::-1]

# 主程式
user_input = input("請輸入文字或句子：")

if is_palindrome(user_input):
    print("這是一個回文！")
else:
    print("這不是一個回文。")
````

````python
# 同時開啟三個檔案：英文、數學（讀取），以及成績單（寫入）
with open("D:/ML/IMLP467/Unit01/data/english_list.csv", "r", encoding="utf-8-sig") as file_e, \
     open("D:/ML/IMLP467/Unit01/data/math_list.csv", "r", encoding="utf-8-sig") as file_m, \
     open("D:/ML/IMLP467/Unit01/Score.csv", "w", encoding="utf-8-sig") as file_out:

    # 1. 讀取所有行並存成清單
    lines_e = file_e.readlines()
    lines_m = file_m.readlines()
         

    # 2. 寫入標題列到新檔案 (如果需要的話)
    file_out.write("姓名,總分\n")

    # 3. 使用 zip 同時抓出兩份檔案的內容 (從 index 1 開始，跳過第一行標題)
    # zip 會把兩份清單像「拉鍊」一樣一對一配對好
    for row_e, row_m in zip(lines_e[1:], lines_m[1:]):
        
        # 處理英文資料：去掉換行、用逗號切開
        data_e = row_e.strip().split(",")
        name = data_e[0]       # 姓名
        eng_score = int(data_e[1]) # 英文分數轉整數

        # 處理數學資料
        data_m = row_m.strip().split(",")
        math_score = int(data_m[1]) # 數學分數轉整數

        # 4. 計算總分
        total = eng_score + math_score

        # 5. 輸出結果
        result = f"{name},{total}"
        print(result)           # 印在螢幕上
        file_out.write(result + "\n") # 寫入檔案並換行
````
